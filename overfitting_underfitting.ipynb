### overfitting(过拟合)：模型在训练集上过分的拟合导致了模型性能很差
### underfitting（欠拟合）：模型在训练集上拟合效果就很差
### 不拟合：模型直接发散

### 过拟合具体表现：

训练过程中训练集（验证集）损失（loss）总体呈下降趋势 ， 准确率（acc）总体呈上升趋势。
验证集（测试集）损失总体呈上升趋势，准确率总体呈下降趋势

### 欠拟合具体表现：

训练过程中训练集（验证集）损失（loss）不呈下降趋势（下降的缓慢）且准确率（acc）不呈上升趋势（上升的缓慢）。

### 直接发散具体表现：

训练过程中损失（loss）出现nan ，inf 且准确率 趋近于0。

### 导致过拟合出现的原因：

最根本的原因是模型无法关注于训练集特征数据之外的的特征

简而言之既是，训练集与验证集（测试集）数据构成（主要在分布、数值两点上）存在不小的差异

其它可能原因：学习率过大、训练epoch过大、模型参数量过大、模型结构太复杂等等
